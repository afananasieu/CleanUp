First of all we have imported our data set with all the variables needed to carry out the service process of our app. After this we created a new data set that will include only the values we need (day, time and intensity) to ask the user what features he wants to request for the service. After this we have created an empty list and through a for loop we have filled it with the elements of the previously created data set. Once this is done, we have generated through inputs questions that will be answered by the user in relation to the variables created in our new data set. Finally we have cosined the answers obtained and the values in the dataset and through the use of the cosine function together with the for loop we will make the users requesting the service and the cleaners reach a common point to carry out the service.

```python
import pandas as pd
import math
import numpy as np
```

```python
##import data set with pandas
dataset_original = pd.read_excel("Cleanup_database.xlsx")
dataset_original
```

```python
## Get a dataset with only the values we need (Day number, Time number & Intensity number)
dataset_trabajo = dataset_original[['Dia_numero', 'Hora_numero', 'intensidad_numero']]
lista = []
for i in range(0,32):
  lista.append(list(dataset_trabajo.iloc[i]))
```

```python
# search question

customer_desired_zone= input("Where do you live? ")
customer_desired_day= (int(input( "Which day do you want to crontract our service?, NOTE:  week=1,  weekend=2.  ")))
customer_desired_hour= (int(input("At what hour do you want to contract our service? NOTE: mañana=1, tarde=2, noche=3, madrugada=4.   ")))
customer_desired_intensity= (int(input("Which intensity do you prefer? NOTE: high=1, Medium=2, Low=3.   ")))
answers_customer=[]
answers_customer.append(customer_desired_day)
answers_customer.append(customer_desired_hour)
answers_customer.append(customer_desired_intensity)

# We cosine the answers obtained
coseno_calculado_answers= np.cos(answers_customer)

# We cosine the values in the dataset
lista_coseno = []
for i in range(0,32):
  lista_coseno.append(np.cos(lista[i]))

# Comparison of cosines
lista_division = []
for i in range(0,32):
  lista_division.append(lista_coseno[i] / coseno_calculado_answers)

perfect_position = []
no_perfect_position = []
no_perfect_position2= []
for i in range(0,32):
  #Case when the elements entered are the same as elements of the datase.
  if sum(lista_division[i]) == 3.:
    perfect_position.append(i)
    # Case when the elements entered are not the same as elements of the dataset. Also we offer the cutomer other options.
  elif 2.5 <= sum(lista_division[i]) <= 3.5:
    no_perfect_position.append(i)
  elif 2.0 <= sum(lista_division[i]) <= 4.0:
    no_perfect_position2.append(i)

for i in range(0, len(perfect_position)):
  print(f" your perfect result is: {dataset_original.iloc[perfect_position[i]]}")

for i in range(0, len(no_perfect_position)):
  print(f"This are not a perfect match for your requirements but these are your most compatible results: {dataset_original.iloc[no_perfect_position[i]]}")

for i in range(0, len(no_perfect_position2)):
  print(f"This are not a perfect match for your requirements but these are your most compatible results:{dataset_original.iloc[no_perfect_position2[i]]}")
```
